```python
# import required modules 
import pandas as pd
from pyAudioAnalysis import audioBasicIO
from pyAudioAnalysis import audioFeatureExtraction
import sys,json, os 
import numpy as np
from pyaudio_featurize import pyaudio_featurize
```

```python
url = "F:/dataset/dataset/test.csv"
predictor_df = pd.read_csv(url)
predictor_df['path'] = predictor_df['path'].apply(lambda s: s[0:])
predictor_df = predictor_df[['path','age']]
predictor_df.dropna(inplace=True)
predictor_df.head()
```

```python
lst = []
extract_labels = pyaudio_featurize('F:/dataset/en/clips/common_voice_en_699711.mp3')
labels = extract_labels[1]
for sample in list(predictor_df['path']):
    features = pyaudio_featurize('F:/dataset/en/clips/%s'%(sample))
    row = list(features[0])
    row.insert(0,sample)
    lst.append(row)
    
labels.insert(0,'path')    
df = pd.DataFrame(lst,columns=labels)
new_df = df.merge(predictor_df,on='path')
new_df.head(10)
```

```python
new_df.to_csv('F:dataset/dataset/audio_testing_data_cleaned.csv') # export cleaned data as csv file
```
